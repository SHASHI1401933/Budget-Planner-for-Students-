```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
import os

class BudgetPlanner:
    """
    A class to manage, track, and visualize personal budget data.
    """
    def __init__(self, month_label=None, currency="₹", savings_goal=None):
        self.records = []  # List to store income/expense dictionaries
        self.currency = currency
        self.savings_goal = savings_goal
        self.month_label = month_label or datetime.now().strftime("%Y-%m")

    def add_entry(self, category, amount, entry_type, date=None):
        """Adds a new income or expense record."""
        entry_type = entry_type.strip().capitalize()
        if entry_type not in ("Income", "Expense"):
            raise ValueError("Type must be 'Income' or 'Expense'")

        amount = float(amount)
        if amount <= 0:
            raise ValueError("Amount must be a positive number")

        self.records.append({
            "date": date or datetime.now().strftime("%Y-%m-%d"),
            "category": category.strip(),
            "amount": amount,
            "type": entry_type
        })

    def to_df(self):
        """Converts the list of records into a pandas DataFrame."""
        return pd.DataFrame(self.records, columns=["date", "category", "amount", "type"])

    def totals(self):
        """Calculates total income, expenses, and savings."""
        df = self.to_df()
        income = df[df["type"] == "Income"]["amount"].sum() if not df.empty else 0.0
        expenses = df[df["type"] == "Expense"]["amount"].sum() if not df.empty else 0.0
        savings = income - expenses
        return income, expenses, savings

    def warn_over_budget(self):
        """Checks if expenses have exceeded income."""
        income, expenses, _ = self.totals()
        return expenses > income

    def expense_by_category(self):
        """Groups total expenses by category."""
        df = self.to_df()
        if df.empty:
            return pd.Series(dtype="float64")

        exp_df = df[df["type"] == "Expense"]
        if exp_df.empty:
            return pd.Series(dtype="float64")

        return exp_df.groupby("category")["amount"].sum().sort_values(ascending=False)

    def goal_status(self):
        """Checks if the savings goal has been met."""
        if self.savings_goal is None:
            return None
        _, _, savings = self.totals()
        return (savings >= self.savings_goal, savings)

    def export_csv(self, filename=None):
        """Exports the budget data to a CSV file."""
        df = self.to_df()
        if filename is None:
            filename = f"budget_{self.month_label}.csv"
        df.to_csv(filename, index=False)
        return os.path.abspath(filename)

    def plot_pie_expenses(self, filename=None):
        """Generates and saves a pie chart of expenses by category."""
        cat_series = self.expense_by_category()
        if cat_series.empty:
            return None

        plt.figure(figsize=(6, 6))
        plt.pie(cat_series.values, labels=cat_series.index, autopct="%1.1f%%", startangle=90)
        plt.title(f"Expenses by Category ({self.month_label})")
        plt.tight_layout()
        outfile = filename or f"expenses_pie_{self.month_label}.png"
        plt.savefig(outfile, dpi=150)
        return os.path.abspath(outfile)

    def plot_bar_income_vs_expense(self, filename=None):
        """Generates and saves a bar chart comparing total income and expenses."""
        income, expenses, _ = self.totals()
        plt.figure(figsize=(5, 4))
        plt.bar(["Income", "Expenses"], [income, expenses], color=["#2e7d32", "#c62828"])
        plt.title(f"Income vs Expenses ({self.month_label})")
        plt.ylabel(f"Amount ({self.currency})")
        for i, v in enumerate([income, expenses]):
            plt.text(i, v, f"{self.currency}{v:,.0f}", ha="center", va="bottom")
        plt.tight_layout()
        outfile = filename or f"income_vs_expense_{self.month_label}.png"
        plt.savefig(outfile, dpi=150)
        return os.path.abspath(outfile)

    def plot_savings_goal_progress(self, filename=None):
        """Generates and saves a bar chart showing progress towards a savings goal."""
        if self.savings_goal is None or self.savings_goal == 0:
            return None

        _, _, savings = self.totals()
        savings_for_plot = max(0, savings) # Avoid negative bar height

        plt.figure(figsize=(5, 4))
        color = '#4caf50' if savings >= self.savings_goal else '#fb8c00'
        plt.bar(["Current Savings", "Goal"], [savings_for_plot, self.savings_goal], color=[color, "#9e9e9e"])
        plt.title(f"Savings Goal Progress ({self.month_label})")
        plt.ylabel(f"Amount ({self.currency})")

        for i, v in enumerate([savings_for_plot, self.savings_goal]):
            plt.text(i, v, f"{self.currency}{v:,.0f}", ha="center", va="bottom")

        progress_pct = (savings / self.savings_goal) * 100
        plt.axhline(y=self.savings_goal, color='grey', linestyle='--', linewidth=1)
        plt.suptitle(f"Progress: {progress_pct:.1f}% of goal", y=0.92, fontsize=10)

        plt.tight_layout(rect=[0, 0, 1, 0.96])
        outfile = filename or f"savings_progress_{self.month_label}.png"
        plt.savefig(outfile, dpi=150)
        return os.path.abspath(outfile)

    def summary_text(self):
        """Generates a multi-line string summarizing the budget."""
        income, expenses, savings = self.totals()
        lines = [
            f"Monthly Income: {self.currency}{income:,.0f}",
            f"Total Expenses: {self.currency}{expenses:,.0f}",
            f"Savings: {self.currency}{savings:,.0f}"
        ]

        if self.warn_over_budget():
            lines.append("Warning: Over Budget (expenses exceed income)")

        if self.savings_goal is not None:
            met, _ = self.goal_status()
            lines.append(f"Savings Goal: {self.currency}{self.savings_goal:,.0f} — {'Met' if met else 'Not Met'}")

        cat_series = self.expense_by_category()
        if not cat_series.empty:
            total_exp = cat_series.sum()
            lines.append("Category-wise Expenses:")
            for cat, amt in cat_series.items():
                pct = (amt / total_exp) * 100 if total_exp > 0 else 0
                lines.append(f"- {cat}: {self.currency}{amt:,.0f} ({pct:.1f}%)")

        return "\n".join(lines)


def interactive_mode():
    """
    Runs the budget planner in an interactive command-line mode.
    """
    print("=== Budget Planner (Interactive Mode) ===")
    month_label = input("Month label (e.g., 2025-08) [default: current month]: ").strip() or None
    currency = input("Currency symbol [default: ₹]: ").strip() or "₹"
    sg_input = input("Savings goal amount (leave blank if none): ").strip()
    savings_goal = float(sg_input) if sg_input else None

    planner = BudgetPlanner(month_label=month_label, currency=currency, savings_goal=savings_goal)

    print("\nEnter entries. Format: Category, Amount, Type(Income/Expense)")
    print("Type 'done' to finish, 'show' to view current summary, or 'help' for examples.\n")

    while True:
        raw_input = input("> ").strip()
        if raw_input.lower() == "done":
            break
        if raw_input.lower() == "help":
            print("Examples:\n  Part-time job, 8000, Income\n  Food, 2000, Expense\n  Books, 1000, Expense")
            continue
        if raw_input.lower() == "show":
            print("\n" + planner.summary_text() + "\n")
            continue
        if not raw_input:
            continue
        try:
            parts = [x.strip() for x in raw_input.split(",")]
            if len(parts) not in (3, 4):
                raise ValueError("Please provide 3 or 4 comma-separated values.")
            category, amount, etype = parts[:3]
            date = parts[3] if len(parts) == 4 else None
            planner.add_entry(category, float(amount), etype, date=date)

            income, expenses, savings = planner.totals()
            print(f"Running Totals — Income: {currency}{income:,.0f}, Expenses: {currency}{expenses:,.0f}, Savings: {currency}{savings:,.0f}")
            if planner.warn_over_budget():
                print("Warning: Over Budget! Expenses exceed income.")
        except Exception as e:
            print(f"Invalid input: {e}")

    # --- Final Output Section ---

    print("\n=== Final Summary ===")
    print(planner.summary_text())

    # Step 1: Generate and display the CSV report immediately.
    print("\n--- Generating CSV Report ---")
    csv_path = planner.export_csv()
    print(f"Exported CSV report to: {csv_path}")
    try:
        print("\n=== CSV Report Content ===")
        with open(csv_path, 'r') as f:
            print(f.read())
    except Exception as e:
        print(f"Could not display CSV content: {e}")

    # Step 2: Generate and display the charts.
    print("\n--- Generating Charts ---")
    pie_path = planner.plot_pie_expenses()
    bar_path = planner.plot_bar_income_vs_expense()
    savings_path = planner.plot_savings_goal_progress()

    if pie_path:
        print(f"Saved pie chart: {pie_path}")
    else:
        print("No expenses found. Pie chart not generated.")
    print(f"Saved bar chart: {bar_path}")
    if savings_path:
        print(f"Saved savings chart: {savings_path}")
    else:
        print("No savings goal set. Savings chart not generated.")

    print("\nDisplaying charts now...")
    plt.show()

    print("\nDone.")


if __name__ == "__main__":
    interactive_mode()
```
